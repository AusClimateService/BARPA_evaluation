```python
import iris
import iris.cube
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
import iris.plot as iplt
```

```python
path = "/short/tp28/eh6215/noAer/"
ct = iris.Constraint(time=lambda t: t.point.month not in  [8,9])
cx = iris.Constraint(longitude=lambda x:85<=x<=210)
cy = iris.Constraint(latitude=lambda y: -55<=y<=15)
```

```python
noAer = iris.load(path+"*noAer*", ct)
iris.util.equalise_attributes(noAer)
noAer
```

```python
radAer = iris.load(path+"*radAer*", ct)
iris.util.equalise_attributes(noAer)
radAer.merge()


```

```python
EA = iris.load(path+"*_EA_*", ct)
iris.util.equalise_attributes(EA)
EA2 = []
for cube in EA:
    if cube.ndim==2:
        EA2.append(cube)
    elif cube.ndim==3:
        for i in range(cube.shape[0]):
            EA2.append(cube[i])
            
EA=iris.cube.CubeList(EA2).merge()
EA
```

```python
r6 = iris.load(path+"*r6*",ct)
iris.util.equalise_attributes(r6)
r6 = r6.merge()
r6


```

```python
era5 = iris.load(path+"*era5*",ct)
iris.util.equalise_attributes(era5)
era5 = era5.merge()
era5 = iris.cube.CubeList([cube.regrid(r6[0],iris.analysis.Linear()) for cube in era5])
```

```python
era5
```

```python
e5 = iris.cube.CubeList()
e5.append(-1*era5.extract_cube("Mean top net short-wave radiation flux") + era5.extract_cube("Mean top downward short-wave radiation flux"))
e5[0].rename("toa_outgoing_shortwave_flux")
e5.append(era5.extract_cube("Mean top downward short-wave radiation flux"))
e5[1].rename("toa_incoming_shortwave_flux")
e5.append(-1*era5.extract_cube("Mean top net long-wave radiation flux"))
e5[2].rename("toa_outgoing_longwave_flux")
e5.append(-1*era5.extract_cube("Mean surface latent heat flux"))
e5[3].rename("surface_upward_latent_heat_flux")
e5.append(-1*era5.extract_cube("Mean surface sensible heat flux"))
e5[4].rename("surface_upward_sensible_heat_flux")
e5.append(era5.extract_cube("Mean surface net short-wave radiation flux"))
e5[5].rename("av_netswsfc")
e5.append(era5.extract_cube("Mean surface net long-wave radiation flux"))
e5[6].rename("surface_net_downward_longwave_flux")
e5.append(era5.extract_cube("Mean surface downward short-wave radiation flux"))
e5[7].rename("surface_downwelling_shortwave_flux_in_air")
```

```python
rad = iris.load(path+"data_312b*.nc",ct&cx&cy)
iris.util.equalise_attributes(rad)
rad=rad.concatenate().merge()

rad.extract_cube("Incoming Solar Flux, Monthly Means").rename("toa_incoming_shortwave_flux")
rad.extract_cube("Top of The Atmosphere Longwave Flux, All-Sky conditions, Monthly Means").rename("toa_outgoing_longwave_flux")
rad.extract_cube("Top of The Atmosphere Shortwave Flux, All-Sky conditions, Monthly Means").rename("toa_outgoing_shortwave_flux")

rad = iris.cube.CubeList([cube.regrid(era5[0],iris.analysis.Linear()) for cube in rad ])
```

```python
# TOA
names = ["toa_outgoing_longwave_flux","toa_outgoing_shortwave_flux"]#,"toa_incoming_shortwave_flux"]
plt.figure(figsize=(24,7))
for i,(data1,data2) in enumerate([(e5,rad),(r6,rad),(EA,rad),(noAer,rad),(radAer,rad)]):
    for j,name in enumerate(names):
        print(name)
        cube = (data1.extract_cube(name).data - data2.extract_cube(name).data).mean(axis=0)
        cube = data1[0][0].copy(data=cube)
        ax=plt.subplot(2,5,i+5*j+1,projection=ccrs.PlateCarree(180))
        iplt.contourf(cube,np.arange(-42.5,43,5),cmap='PiYG')
        plt.title(['era5 - obs\n','r6 -obs\n','EA-obs\n', 'noEA - obs\n','radEA-obs\n',"noEA - r6\n",'radEA - r6\n'][i]+name)
        plt.colorbar()
        ax.coastlines()
```

```python
names = ["surface_upward_latent_heat_flux","surface_upward_sensible_heat_flux","av_netswsfc","surface_net_downward_longwave_flux"]#,"surface_downwelling_shortwave_flux_in_air"]#,"toa_incoming_shortwave_flux"]
plt.figure(figsize=(20,12))

lon = r6[0].coord('longitude').points
lat = r6[0].coord('latitude').points

for i,(data1,data2) in enumerate([(r6,e5),(EA,e5),(noAer,e5),(radAer,e5)]):
    for j,name in enumerate(names):
        cube = (data1.extract_cube(name).data - data2.extract_cube(name).data).mean(axis=0)
        cube = data1[0][0].copy(data=cube)
        #cube = data1.extract_cube(name).collapsed('time',iris.analysis.MEAN)
        ax=plt.subplot(4,4,i+4*j+1,projection=ccrs.PlateCarree(180))
        iplt.contourf(cube,np.arange(-32.5,33,5),cmap='PiYG',extend='both')
        plt.title(['r6 - era5\n', 'EA - era5\n','noEA - era5\n',"radEA - era5\n", 'noEA-r6','radEA-r6'][i]+name)
        #plt.colorbar()#ticks=np.arange(-30,31,10))
        ax.coastlines()
        
```

```python
net = {'r6':{},"noEA":{},'radEA':{},'era5':{}}

time=r6[0].coord('time')
for cube in r6+noAer+radAer:
    cube.remove_coord('time')
    cube.add_dim_coord(time,0)
    try:
        cube.remove_coord("forecast_period")
    except:
        1

cx=iris.Constraint(longitude=lambda x: x<175)
net['r6']['net'] = (r6.extract("surface_net_downward_longwave_flux")[0]+r6.extract("av_netswsfc")[0]-r6.extract("surface_upward_latent_heat_flux")[0]-r6.extract("surface_upward_sensible_heat_flux")[0]).extract(cx)
net['noEA']['net'] = (noAer.extract("surface_net_downward_longwave_flux")[0]+noAer.extract("av_netswsfc")[0]-noAer.extract("surface_upward_latent_heat_flux")[0]-noAer.extract("surface_upward_sensible_heat_flux")[0]).extract(cx)
net['radEA']['net'] = (radAer.extract("surface_net_downward_longwave_flux")[0]+radAer.extract("av_netswsfc")[0]-radAer.extract("surface_upward_latent_heat_flux")[0]-radAer.extract("surface_upward_sensible_heat_flux")[0]).extract(cx)
net['era5']['net'] = (e5.extract("surface_net_downward_longwave_flux")[0]+e5.extract("av_netswsfc")[0]-e5.extract("surface_upward_latent_heat_flux")[0]-e5.extract("surface_upward_sensible_heat_flux")[0]).extract(cx)
```
