```python
import iris
import numpy as np
```

```python
path = "/g/data/tp28/dev/eh6215/mon_pl/"
trials = ['r1','r2','r4','r6']
yearmon  = [(2012,10),(2012,11),(2012,12),(2013,1),(2013,2),(2013,3),(2013,4),(2013,5),(2013,6),(2013,7),(2013,8)]
e5_vars = ['z','u','v','q','t','w']
e5_path = "/g/data/rt52/era5/pressure-levels/monthly-averaged/{var}/{year}/{var}_era5_moda_pl_{year}{mon:02d}01-{year}{mon:02d}*.nc"

cp=iris.Constraint(pressure_level = lambda p: np.any([(np.abs(p.point-pp)<1) for pp in [200,500,850,1000]]))
cp500=iris.Constraint(pressure_level = lambda p: np.any([(np.abs(p.point-pp)<1) for pp in [500]]))


```

```python
data={}
for trial in trials:
    print(trial)
    data[trial]=iris.load([path+"%s_%04d%02d*.nc"%(trial,year,mon) for year,mon in yearmon])
    iris.util.equalise_attributes(data[trial])
    data[trial]=data[trial].concatenate()
```

```python
e5paths = [[e5_path.format(year=year,mon=month,var=v) for v in e5_vars] for (year,month) in yearmon]
e5paths = [item for sublist in e5paths for item in sublist]

data['era5'] = iris.load(e5paths, cp)
iris.util.equalise_attributes(data['era5'])
data['era5'] = iris.cube.CubeList([cube.regrid(data['r1'][0],iris.analysis.Linear()) for cube in data['era5'].concatenate()])


```
